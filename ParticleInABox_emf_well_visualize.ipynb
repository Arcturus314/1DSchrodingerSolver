```python
import matplotlib.pyplot as plt
import numpy as np
import math

fig, axarr = plt.subplots(3, sharex=True, figsize=(3,5))

m = 1
hbar = 1
pi = 3.1415926
num_pts = 100
energy_scaling_factor = 12

well_bounds = [0,1]
l = well_bounds[1] - well_bounds[0]

def psi(n,x,l):
    #print(str(x) + " " + str(n*pi*x/l))
    return math.sin(n*pi*x/l)

def energy(n,m,l):
    return n**2 * hbar**2 * pi**2 / (2 * m * l**2)

psi_xlist  = np.linspace(0,1,num_pts)

psi_nlist  = [1,2,3]
linestyles = ['--', '-.', ':']

psi_ylists  = []
energy_list = []
energy_lists = []

for n in psi_nlist:
    psi_ylists.append([])
    energy_lists.append([])
    energy_list.append(energy(n,m,l))
    for x in psi_xlist:
        psi_ylists[-1].append(psi(n, x, l)+energy_list[-1]/energy_scaling_factor)
        energy_lists[-1].append(energy_list[-1]/energy_scaling_factor)
        
#for i in [0,2]:
#    for n in psi_nlist:
#        axarr[i].plot(psi_xlist, psi_ylists[n-1], c='black', linestyle=linestyles[n-1])
#        axarr[i].plot(psi_xlist, energy_lists[n-1], c='black', alpha = 0.3)

#plotting well
for i in [2]:
    axarr[i].plot([0,l], [max(psi_ylists[-1]),max(psi_ylists[-1])-l], c='black')
    axarr[i].arrow(0,max(psi_ylists[-1]),0,-max(psi_ylists[-1])+0.7*l,head_width=0.03, head_length=0.1, fc='black')
    axarr[i].arrow(l,max(psi_ylists[-1])-l,0,-max(psi_ylists[-1])+1.7*l, head_width=0.03, head_length=0.1, fc='black')
    
    #axarr[i].plot([0,l], [max(psi_ylists[-1]),max(psi_ylists[-1])], c='black')
    #axarr[i].arrow(l,max(psi_ylists[-1]),0,-max(psi_ylists[-1])+0.7*l, head_width=0.03, head_length=0.1, fc='black')
    #axarr[i].arrow(0,max(psi_ylists[-1]),0,-max(psi_ylists[-1])+0.7*l, head_width=0.03, head_length=0.1, fc='black')

    axarr[i].text(-0.03,0, "∞")
    axarr[i].text(l-0.03,0, "∞")
    
for i in [0]:
    axarr[i].plot([0,l], [l,0], c='black')
    axarr[i].arrow(l,0,0,max(psi_ylists[-1])-0.7*l,head_width=0.03, head_length=0.1, fc='black')
    axarr[i].arrow(0,l,0,max(psi_ylists[-1])-1.7*l, head_width=0.03, head_length=0.1, fc='black')
    
    #axarr[i].plot([0,l], [0,0], c='black')
    #axarr[i].arrow(l,0,0,max(psi_ylists[-1])-0.7*l,head_width=0.03, head_length=0.1, fc='black')
    #axarr[i].arrow(0,0,0,max(psi_ylists[-1])-0.7*l, head_width=0.03, head_length=0.1, fc='black')
    
    axarr[i].text(l-0.03,max(psi_ylists[-1])-0.4*l, "∞")
    axarr[i].text(0-0.03,max(psi_ylists[-1])-0.4*l, "∞")
    
for i in [0,2]:
    axarr[i].set_xlim([-0.1,1.1])
    axarr[i].set_ylim([-0.5, max(psi_ylists[-1])*1.2])
#    axarr[i].set_ylabel("Potential")
    
for i in range(3):
    axarr[i].spines['top'].set_visible(False)
    axarr[i].spines['right'].set_visible(False)
    axarr[i].spines['bottom'].set_visible(False)
    axarr[i].spines['left'].set_visible(False)
    axarr[i].set_yticks([])
    axarr[i].set_xticks([])
#    axarr[i].set_xlabel("x Position")

axarr[1].set_ylim([-1,1])
axarr[1].text(0.2,0,"1.43eV Band Gap")
axarr[1].arrow(0.5,-1,0,0.8, head_width=0.03, head_length=0.1, fc='black')
axarr[1].arrow(0.5,1,0,-0.7, head_width=0.03, head_length=0.1, fc='black')
axarr[1].plot([0.3,0.7], [1,1], c="black", linewidth=3)
axarr[1].plot([0.3,0.7], [-1,-1], c="black", linewidth=5)


#plt.xlabel("x Position")
axarr[1].set_ylabel("Energy")
plt.subplots_adjust(hspace=0)
plt.show()
    
        
```
